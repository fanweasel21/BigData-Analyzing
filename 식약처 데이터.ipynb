```python
exc = ['학교', '사회복지시설','병원', '산업체', '아시아식', '공공기관', '기숙사', '출장조리', '유원지', '어린이집',
        '기타 집단급식소', '백화점', '수련원', '키즈카페', '극장', '산후조리원', '도매업(유통)', '방문판매']
```

```python
df_1 = df_1.dropna()
df_1.shape
```

```python
df_2 = df_1.copy()
df_2
```

```python
# 컬럼 순서 정리
df_2 = df_2[['area', 'year', 'category']]
```

```python
df_2
```

```python
grouped_df = df_2.groupby(['area', 'year', 'category'])
```

```python
new_df = pd.DataFrame(grouped_df.size().reset_index(name = ""))
new_df
```

```python
# 연도 이상치 제거
odd = ['5001', '1194', '9960', '1881', '1883', '1889', 
       '1897', '1899', '1909', '1916', '1197', '1910', '1913']
```

```python
for j in range(len(odd)):
    odd_year = new_df[new_df['year'] == odd[j]].index
    new_df = new_df.drop(odd_year)
```

```python
new_df
```

```python
# pivot table
dfdf = new_df.pivot(index =['area', 'year'], columns ='category')
dfdf
```

```python
# NaN값을 0으로 변경
dfdf = dfdf.fillna(0)
dfdf
```

```python
# 소수점을 정수로 변경
dfdf = dfdf.astype(int)
dfdf
```
